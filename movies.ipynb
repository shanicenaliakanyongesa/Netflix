```python
import seaborn as sns

import matplotlib.pyplot as plt

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

# 1. Distribution of content type
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

data['type'].value_counts().plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of Content Type')
axes[0, 0].set_xlabel('Type')
axes[0, 0].set_ylabel('Count')

# 2. Top 10 countries with most content
data['country'].str.split(', ').explode().value_counts().head(10).plot(kind='barh', ax=axes[0, 1], color='lightcoral')
axes[0, 1].set_title('Top 10 Countries with Most Content')
axes[0, 1].set_xlabel('Count')

# 3. Rating distribution
data['rating'].value_counts().plot(kind='bar', ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Distribution of Ratings')
axes[1, 0].set_xlabel('Rating')
axes[1, 0].set_ylabel('Count')
axes[1, 0].tick_params(axis='x', rotation=45)

# 4. Release year distribution
data['release_year'].value_counts().sort_index().plot(kind='line', ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Content by Release Year')
axes[1, 1].set_xlabel('Release Year')
axes[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()

# Additional analysis: Duration distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Movies duration
movies_duration = data[data['type'] == 'Movie']['duration'].str.extract('(\d+)').astype(float)
movies_duration.plot(kind='hist', bins=30, ax=axes[0], color='orange', edgecolor='black')
axes[0].set_title('Distribution of Movie Duration (minutes)')
axes[0].set_xlabel('Duration (minutes)')
axes[0].set_ylabel('Count')

# TV Shows seasons
tv_seasons = data[data['type'] == 'TV Show']['duration'].str.extract('(\d+)').astype(float)
tv_seasons.plot(kind='hist', bins=20, ax=axes[1], color='teal', edgecolor='black')
axes[1].set_title('Distribution of TV Show Seasons')
axes[1].set_xlabel('Number of Seasons')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
%pip install scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Create a simple movie recommendation system based on content similarity


# Filter only movies
movies = data[data['type'] == 'Movie'].reset_index(drop=True)

# Combine relevant features for similarity calculation
movies['content_features'] = movies['listed_in'] + ' ' + movies['description'].fillna('')

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = vectorizer.fit_transform(movies['content_features'])

# Calculate cosine similarity matrix
similarity_matrix = cosine_similarity(tfidf_matrix)

# Function to get movie recommendations
def recommend_movies(movie_title, n_recommendations=5):
    """
    Recommend similar movies based on title
    """
    try:
        # Find movie index
        movie_idx = movies[movies['title'].str.lower() == movie_title.lower()].index[0]
        
        # Get similarity scores
        similarity_scores = list(enumerate(similarity_matrix[movie_idx]))
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
        
        # Get top recommendations (excluding the movie itself)
        recommendations = similarity_scores[1:n_recommendations+1]
        
        print(f"\nMovies similar to '{movies.loc[movie_idx, 'title']}':")
        print("-" * 80)
        for idx, (movie_idx_rec, score) in enumerate(recommendations, 1):
            rec_movie = movies.loc[movie_idx_rec]
            print(f"{idx}. {rec_movie['title']} ({rec_movie['release_year']}) - Similarity: {score:.2%}")
            print(f"   Rating: {rec_movie['rating']} | Duration: {rec_movie['duration']}")
            print(f"   Genre: {rec_movie['listed_in']}\n")
        
        return [movies.loc[idx, 'title'] for idx, _ in recommendations]
    
    except IndexError:
        print(f"Movie '{movie_title}' not found in database")
        return []

# Example usage
recommend_movies('Inception', n_recommendations=5)
```

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# -----------------------------
# Initialize Dash App
# -----------------------------
app = dash.Dash(__name__)

# -----------------------------
# Layout
# -----------------------------
app.layout = html.Div([

    html.H1("Netflix Content Dashboard",
            style={
                'textAlign': 'center', 
                'marginBottom': 40,
                'color': '#E50914',
                'fontFamily': 'Arial, sans-serif',
                'fontSize': '48px'
            }),

    # Filters Section
    html.Div([
        html.Div([
            html.Label("Select Content Type:", 
                      style={'fontWeight': 'bold', 'marginBottom': '8px', 'display': 'block'}),
            dcc.Dropdown(
                id='type-dropdown',
                options=[{'label': 'All', 'value': 'All'}] +
                        [{'label': str(t), 'value': str(t)} for t in sorted(data['type'].dropna().unique())],
                value='All',
                style={'width': '100%'}
            )
        ], style={'width': '45%', 'display': 'inline-block', 'marginRight': '5%'}),

        html.Div([
            html.Label("Select Rating:", 
                      style={'fontWeight': 'bold', 'marginBottom': '8px', 'display': 'block'}),
            dcc.Dropdown(
                id='rating-dropdown',
                options=[{'label': 'All', 'value': 'All'}] +
                        [{'label': str(r), 'value': str(r)} for r in sorted(data['rating'].dropna().unique())],
                value='All',
                style={'width': '100%'}
            )
        ], style={'width': '45%', 'display': 'inline-block'})

    ], style={
        'marginBottom': 40, 
        'padding': '20px',
        'backgroundColor': '#f8f9fa',
        'borderRadius': '10px'
    }),

    # Row 1: Two graphs side by side
    html.Div([
        html.Div([
            dcc.Graph(id='content-type-chart', style={'height': '400px'})
        ], style={'width': '48%', 'display': 'inline-block', 'verticalAlign': 'top'}),
        
        html.Div([
            dcc.Graph(id='top-countries-chart', style={'height': '400px'})
        ], style={'width': '48%', 'display': 'inline-block', 'marginLeft': '4%', 'verticalAlign': 'top'})
    ], style={'marginBottom': 30}),

    # Row 2: Two graphs side by side
    html.Div([
        html.Div([
            dcc.Graph(id='release-year-chart', style={'height': '400px'})
        ], style={'width': '48%', 'display': 'inline-block', 'verticalAlign': 'top'}),
        
        html.Div([
            dcc.Graph(id='rating-distribution-chart', style={'height': '400px'})
        ], style={'width': '48%', 'display': 'inline-block', 'marginLeft': '4%', 'verticalAlign': 'top'})
    ], style={'marginBottom': 30}),

    # Row 3: Full width graph
    html.Div([
        dcc.Graph(id='duration-chart', style={'height': '400px'})
    ], style={'marginBottom': 30})

], style={
    'fontFamily': 'Arial, sans-serif',
    'padding': '20px 40px',
    'backgroundColor': '#ffffff',
    'maxWidth': '1400px',
    'margin': '0 auto'
})

# -----------------------------
# Callback
# -----------------------------
@app.callback(
    [Output('content-type-chart', 'figure'),
     Output('top-countries-chart', 'figure'),
     Output('release-year-chart', 'figure'),
     Output('rating-distribution-chart', 'figure'),
     Output('duration-chart', 'figure')],
    [Input('type-dropdown', 'value'),
     Input('rating-dropdown', 'value')]
)
def update_dashboard(selected_type, selected_rating):
    
    try:
        print(f"\n=== Callback triggered ===")
        print(f"selected_type: {selected_type} (type: {type(selected_type)})")
        print(f"selected_rating: {selected_rating} (type: {type(selected_rating)})")
        
        filtered_data = data.copy()
        print(f"Initial data shape: {filtered_data.shape}")

        # Apply filters with proper type handling
        if selected_type != 'All' and selected_type is not None:
            print(f"Filtering by type: {selected_type}")
            filtered_data = filtered_data[filtered_data['type'].astype(str) == str(selected_type)]
            print(f"After type filter: {filtered_data.shape}")

        if selected_rating != 'All' and selected_rating is not None:
            print(f"Filtering by rating: {selected_rating}")
            filtered_data = filtered_data[filtered_data['rating'].astype(str) == str(selected_rating)]
            print(f"After rating filter: {filtered_data.shape}")

        # If no data after filtering
        if filtered_data.empty:
            empty_fig = go.Figure()
            empty_fig.add_annotation(text="No data available",
                                     x=0.5, y=0.5,
                                     showarrow=False)
            return empty_fig, empty_fig, empty_fig, empty_fig, empty_fig

        # 1️⃣ Content Type Distribution
        type_counts = filtered_data['type'].value_counts().reset_index()
        type_counts.columns = ['type', 'count']

        fig1 = px.bar(type_counts,
                      x='type',
                      y='count',
                      title='Content Type Distribution',
                      color='type',
                      color_discrete_sequence=['#E50914', '#B20710'])
        fig1.update_layout(
            showlegend=False,
            title_font_size=18,
            title_x=0.5,
            plot_bgcolor='rgba(0,0,0,0)',
            paper_bgcolor='rgba(0,0,0,0)'
        )

        # 2️⃣ Top 10 Countries
        countries_series = filtered_data['country'].dropna().astype(str)
        countries_series = countries_series[countries_series != '']
        
        if not countries_series.empty:
            top_countries = (
                countries_series
                .str.split(', ')
                .explode()
                .value_counts()
                .head(10)
                .reset_index()
            )
            top_countries.columns = ['country', 'count']

            fig2 = px.bar(top_countries,
                          x='count',
                          y='country',
                          title='Top 10 Countries',
                          orientation='h',
                          color='count',
                          color_continuous_scale=['#B20710', '#E50914'])
            fig2.update_yaxes(categoryorder='total ascending')
            fig2.update_layout(
                showlegend=False,
                title_font_size=18,
                title_x=0.5,
                plot_bgcolor='rgba(0,0,0,0)',
                paper_bgcolor='rgba(0,0,0,0)'
            )
        else:
            fig2 = go.Figure()
            fig2.add_annotation(text="No country data available",
                                x=0.5, y=0.5,
                                showarrow=False)

        # 3️⃣ Release Year Trend
        year_counts = (
            filtered_data['release_year']
            .dropna()
            .value_counts()
            .sort_index()
            .reset_index()
        )
        year_counts.columns = ['release_year', 'count']

        fig3 = px.line(year_counts,
                       x='release_year',
                       y='count',
                       title='Content by Release Year',
                       markers=True)
        fig3.update_traces(line_color='#E50914', marker=dict(size=6))
        fig3.update_layout(
            title_font_size=18,
            title_x=0.5,
            plot_bgcolor='rgba(0,0,0,0)',
            paper_bgcolor='rgba(0,0,0,0)',
            hovermode='x unified'
        )

        # 4️⃣ Rating Distribution
        rating_counts = filtered_data['rating'].dropna().value_counts().reset_index()
        rating_counts.columns = ['rating', 'count']

        fig4 = px.bar(rating_counts,
                      x='rating',
                      y='count',
                      title='Rating Distribution',
                      color='count',
                      color_continuous_scale=['#B20710', '#E50914'])
        fig4.update_layout(
            showlegend=False,
            title_font_size=18,
            title_x=0.5,
            plot_bgcolor='rgba(0,0,0,0)',
            paper_bgcolor='rgba(0,0,0,0)'
        )

        # 5️⃣ Movie Duration Distribution
        movie_data = filtered_data[filtered_data['type'] == 'Movie']

        if not movie_data.empty and 'duration' in movie_data.columns:
            duration_data = (
                movie_data['duration']
                .dropna()
                .astype(str)
                .str.extract(r'(\d+)', expand=False)
            )
            duration_data = pd.to_numeric(duration_data, errors='coerce')
            duration_data = duration_data.dropna()

            if not duration_data.empty:
                fig5 = px.histogram(duration_data,
                                    nbins=30,
                                    title='Movie Duration Distribution (minutes)',
                                    labels={'value': 'Duration (min)', 'count': 'Number of Movies'})
                fig5.update_traces(marker_color='#E50914')
                fig5.update_layout(
                    title_font_size=18,
                    title_x=0.5,
                    plot_bgcolor='rgba(0,0,0,0)',
                    paper_bgcolor='rgba(0,0,0,0)',
                    bargap=0.1
                )
            else:
                fig5 = go.Figure()
                fig5.add_annotation(text="No valid duration data",
                                    x=0.5, y=0.5,
                                    showarrow=False)
        else:
            fig5 = go.Figure()
            fig5.add_annotation(text="No movies in selection",
                                x=0.5, y=0.5,
                                showarrow=False)

        return fig1, fig2, fig3, fig4, fig5
    
    except Exception as e:
        print(f"ERROR in callback: {type(e).__name__}: {str(e)}")
        import traceback
        traceback.print_exc()
        
        # Return empty figures on error
        empty_fig = go.Figure()
        empty_fig.add_annotation(text=f"Error occurred - check console",
                                 x=0.5, y=0.5,
                                 showarrow=False)
        return empty_fig, empty_fig, empty_fig, empty_fig, empty_fig


# -----------------------------
# Run App
# -----------------------------
app.run(debug=True)
```
